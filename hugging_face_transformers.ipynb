```python
from transformers import pipeline
```

```python
clf = pipeline("sentiment-analysis")
```

```python
print(clf("I love Hugging Face!"))
```

```python
gen = pipeline("text-generation", model="gpt2")
```

```python
print(gen("Once upon a time", max_length=50))
```

```python
qa = pipeline("question-answering")
result = qa(
    question="What is Hugging Face?",
    context="Hugging Face is a company that builds NLP tools."
)
print(result)  # {'answer': 'a company that builds NLP tools', ...}
```

```python
ner = pipeline("ner", grouped_entities=True)
print(ner("My name is John and I live in New York."))
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

article = "Long article text here..."
inputs = tokenizer.encode("summarize: " + article, return_tensors="pt", max_length=512, truncation=True)

summary_ids = model.generate(inputs, max_length=60, min_length=30, do_sample=False)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(summary)
```
